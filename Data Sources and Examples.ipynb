## Art Galleries, Museums, and Live Theater
### We want to provide the user with the venues for art galleries, museums, and live theater.  These all have permanent venues that can be located on a map.  Other art and culture activities often use various venues such as churches and schools so we cannot provide a definite location for them.  We intend to set up a database for those activities without a home but that is outside the scope of this project.

## 2. Foursquare Data
### I downloaded venue data from Foursquare and created dataframes for all eight community regions.  The columns selected were
filtered_columns = \['name', 'address', 'city', 'category', 'lat', 'lng'\]

### These eight dataframes were concatenated into one dataframe with 357 rows. 
### Because the radius was set wide enough to get all of the rural areas there was some overlap of the radius and also some Toronto venues were picked up. The data was processed to remove duplicate rows and any venues with Toronto as the city. 
### This left 329 venues. There were 97 unique categories included in the dataframe.  After looking at the list of categories from the dataframe I selected these three relevant categories: Art gallery, Museum, and Theater.
